```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
def get_headline_urls(page_number):
    base_url = f"https://www.bbc.com/urdu/topics/c40379e2ymxt?page={page_number}"
    response = requests.get(base_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    headline_urls = []
    headlines = soup.find_all('h2')
    for headline in headlines:
        link = headline.find('a')['href']
        headline_urls.append(link)
    return headline_urls

def get_articles(url):
    base_url = url
    response = requests.get(base_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    articles = soup.find_all('p')
    return [article.text.strip() for article in articles]

def save_all_headlines_to_csv():
    total_headlines = 0
    with open('bbc_science_headlines.csv', 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(["Page", "Headlineno.", "Article", "Topic"])

        for page_number in range(1, 41):
            headline_urls = get_headline_urls(page_number)
#             print("got the urls.....")
            for i, url in enumerate(headline_urls, start=1):
                new_articles = get_articles(url)
#                 print("getting articles.......")
                for headline in new_articles:
                    total_headlines += 1
#                     print("on article ", i)
                    writer.writerow([page_number, i, headline, "Science"])
            print(f"Articles from page {page_number} saved to bbc_science_headlines.csv")

    print("All headlines saved to bbc_science_headlines.csv")

if __name__ == "__main__":
    save_all_headlines_to_csv()
```

```python
class HeadlineScraper:
    def __init__(self, base_url, output_file, topic):
        self.base_url = base_url
        self.output_file = output_file
        self.topic = topic

    def get_headlines(self):
        response = requests.get(self.base_url)
        soup = BeautifulSoup(response.content, 'html.parser')
        headlines = soup.find_all('h2')
        return [headline.text.strip() for headline in headlines]

    def save_to_csv(self, headlines):
        with open(self.output_file, 'w', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerow(["Index", "Headline", "Topic"])

            for i, headline in enumerate(headlines, start=1):
                writer.writerow([i, headline, self.topic])

    def scrape_and_save(self):
        new_headlines = self.get_headlines()
        self.save_to_csv(new_headlines)
        print(f"Headlines saved to {self.output_file}")

if __name__ == "__main__":
    base_url = "https://www.dawnnews.tv/latest-news"
    output_file = "dawn_latest_headlines.csv"
    topic = "Latest"
    scraper = HeadlineScraper(base_url, output_file, topic)
    scraper.scrape_and_save()
```
